```python
import tkinter as tk
import random
import math
import time
from PIL import Image, ImageDraw
class Bird:
    def __init__(self,name):
        self.x = random.uniform(100,1000)
        self.y = random.uniform(100,800)
        self.vx = random.uniform(-1.5,1.5)
        self.vy = random.uniform(-1.5,1.5)
        self.ax = 0
        self.ay = 0
        self.name = name
        self.color = "blue"
        self.maxSpeed = 1.5
        self.random_number = random.uniform(0.7, 1.3)
        self.radius = 150 * self.random_number
        self.alive=True
        self.hungry = 100
        self.hungry_state = False if self.hungry > 30 else True

        self.edge_buffer = 100
        self.energy = 100  
        self.maxEnergy = 100
        
    
                    
    def draw(self,canvas):
        size = 6
        baseSize = 3
        angle = math.atan2(self.vy, self.vx)
        point1 = [self.x + size*math.cos(angle), self.y + size*math.sin(angle)]
        point2 = [self.x - size*math.cos(angle) + baseSize*math.cos(angle+math.pi/2.0), \
                  self.y - size*math.sin(angle) + baseSize*math.sin(angle+math.pi/2.0) ]
        point3 = [self.x - size*math.cos(angle) - baseSize*math.cos(angle+math.pi/2.0), \
                  self.y - size*math.sin(angle) - baseSize*math.sin(angle+math.pi/2.0) ]
        canvas.create_polygon(point1+point2+point3,fill=self.color,tags=self.name)
        
    
    def move(self, canvas):
        distance = 5
        if self.vx == 0.0:
            self.vx = 0.0000001
        if self.vy == 0.0:
            self.vy = 0.0000001
        unitVX = self.vx / math.sqrt(self.vx**2 + self.vy**2)
        unitVY = self.vy / math.sqrt(self.vx**2 + self.vy**2)
        self.x += self.vx if math.sqrt(self.vx**2+self.vy**2) < self.maxSpeed else unitVX*self.maxSpeed
        self.y += self.vy if math.sqrt(self.vx**2+self.vy**2) < self.maxSpeed else unitVY*self.maxSpeed
        self.x = self.x%1400.0
        self.y = self.y%800.0
        self.vx += self.ax
        self.vy += self.ay 
        self.hungry -= 0.1
        if self.maxSpeed>1.5:
            self.maxSpeed -= 0.002
        if self.hungry< 30 and self.maxSpeed>0.7:
            self.maxSpeed -= 0.002

        if self.alive == False:
            self.x= -999
            self.y= -999
        if self.alive:
            canvas.delete(self.name)
            self.draw(canvas)
        else:
            canvas.delete(self.name)


        
    def distance(self, otherBird):
        base_distance = math.sqrt( (self.x-otherBird.x)*(self.x-otherBird.x) + \
                          (self.y-otherBird.y)*(self.y-otherBird.y))
        return base_distance * self.random_number
foods=0
food_position=[]
class Island:
    def __init__(self,name, canvas, n, radius, x, y):
        self.canvas = canvas
        self.x = x
        self.y = y
        self.points = [] 
        self.edges = []
        self.radius = radius
        
    def generate_food(self,canvas):
        global foods,food_position
        canvas.after(50,self.generate_food,canvas)
        if  foods < 20:
            # print(foods)
            food_x = self.x + random.uniform(-self.radius/3*2, self.radius/3*2)
            food_y = self.y + random.uniform(-self.radius/3*2, self.radius/3*2)
            food_id =self.canvas.create_oval(food_x-2, food_y-2, food_x+2, food_y+2, fill='white')
            food_position.append((food_x, food_y, food_id))
            foods += 1

    def draw(self, canvas, n, radius, x, y):
        angle = 2 * math.pi / n   
        for i in range(n):
            radius = radius + random.uniform(-radius * 0.2, radius * 0.2) 
            x = self.x + radius * math.cos(i * angle)
            y = self.y + radius * math.sin(i * angle)
            self.points.append((x, y))
            if i > 0:
                self.edges.append((self.points[i - 1], self.points[i]))
        # Connect the last point to the first point to close the polygon
        self.edges.append((self.points[-1], self.points[0]))
        self.canvas.create_polygon([p for point in self.points for p in point], fill='sandybrown', outline='black', width=2)


import random
import math

class Predator:
    def __init__(self, name,x,y):
        self.x = x+random.uniform(10, 10)
        self.y = y+random.uniform(10, 10)
        self.vx = random.uniform(-2, 2)  
        self.vy = random.uniform(-2, 2)
        self.ax = 0
        self.ay = 0
        self.name = name
        self.color = "black"  
        self.maxSpeed = 1.5  
        self.random_number = random.uniform(0.7, 1.3)
        self.radius = 400 * self.random_number
        self.hungry = 100
        self.edge_buffer = 100
        self.energy = 100
        self.maxEnergy = 100
    

    def draw(self,canvas):
        size = 8
        baseSize = 4
        angle = math.atan2(self.vy, self.vx)
        point1 = [self.x + size*math.cos(angle), self.y + size*math.sin(angle)]
        point2 = [self.x - size*math.cos(angle) + baseSize*math.cos(angle+math.pi/2.0), \
                  self.y - size*math.sin(angle) + baseSize*math.sin(angle+math.pi/2.0) ]
        point3 = [self.x - size*math.cos(angle) - baseSize*math.cos(angle+math.pi/2.0), \
                  self.y - size*math.sin(angle) - baseSize*math.sin(angle+math.pi/2.0) ]
        canvas.create_polygon(point1+point2+point3,fill=self.color,tags=self.name)

    def move(self, canvas):
       
        if self.vx == 0.0:
            self.vx = 0.0000001
        if self.vy == 0.0:
            self.vy = 0.0000001
        unitVX = self.vx / math.sqrt(self.vx**2 + self.vy**2)
        unitVY = self.vy / math.sqrt(self.vx**2 + self.vy**2)
        self.x += self.vx if math.sqrt(self.vx**2+self.vy**2) < self.maxSpeed else unitVX*self.maxSpeed
        self.y += self.vy if math.sqrt(self.vx**2+self.vy**2) < self.maxSpeed else unitVY*self.maxSpeed
        self.x = self.x%1400.0
        self.y = self.y%800.0
        self.vx += self.ax
        self.vy += self.ay    
        speed = math.sqrt(self.vx**2 + self.vy**2)
        if speed > self.maxSpeed:
            self.vx = self.vx / speed * self.maxSpeed
            self.vy = self.vy / speed * self.maxSpeed
        if self.maxSpeed>1.5:
            self.maxSpeed -= 0.001
        canvas.delete(self.name)
        self.draw(canvas)

    def distance(self, other):
        base_distance = math.sqrt((self.x - other.x) ** 2 + (self.y - other.y) ** 2)
        return base_distance * self.random_number

    def find_nearest_bird(self, bird_list):
        min_distance = float('inf')
        nearest_bird = None
        for bird in bird_list:
            distance = self.distance(bird)
            if distance < min_distance:
                min_distance = distance
                nearest_bird = bird
        return nearest_bird, min_distance


def initialise(window):
    window.resizable(False,False)
    canvas = tk.Canvas(window,width=1400,height=800,bg='lightblue')
    canvas.pack()
    return canvas

def is_point_in_polygon(point, polygon):
    x, y = point
    n = len(polygon)
    inside = False
    p1x, p1y = polygon[0]
    for i in range(n + 1):
        p2x, p2y = polygon[i % n]
        if y > min(p1y, p2y):
            if y <= max(p1y, p2y):
                if x <= max(p1x, p2x):
                    if p1y != p2y:
                        xinters = (y - p1y) * (p2x - p1x) / (p2y - p1y) + p1x
                    if p1x == p2x or x <= xinters:
                        inside = not inside
        p1x, p1y = p2x, p2y
    return inside


def addBirds(canvas):
    birdList = []
    noOfBirds = 60
    for n in range(noOfBirds):
        bird = Bird("bird"+str(n))
        birdList.append(bird)
        bird.draw(canvas)
    return birdList

def addPredators(canvas):
    PredatorsList = []    
    predator1 = Predator("Predators1",600,400)
    PredatorsList.append(predator1)
    
    predator1.draw(canvas)
    
    return PredatorsList


def addIsland(canvas):
    IslandList=[]
    island1 = Island("Island1",canvas,20, 100, 700,400)
    # island2 = Island("Island2",canvas,20, 100, 1200,700)

    IslandList.append(island1)
    # IslandList.append(island2)
    for island in IslandList:
        island.generate_food(canvas)
    island1.draw(canvas,20, 150, 200,200)
    # island2.draw(canvas,10, 100, 800,800)

    return IslandList


def draw_red_x(canvas, x, y):
    size=5
    left_x = x - size
    right_x = x + size
    top_y = y - size
    bottom_y = y + size
    canvas.create_line(left_x, top_y, right_x, bottom_y, fill='red', width=2)
    canvas.create_line(left_x, bottom_y, right_x, top_y, fill='red', width=2)
    
def movePredator(listOfPredators, canvas, birdList,islands):
    
    
    showNeighboursP(canvas,birdList,islands,listOfPredators)    


    for predator in listOfPredators:
        predator.move(canvas)
        predator.hungry -= 0.1
        

    for predator in listOfPredators:
        
        
        forceXcoherence=0
        forceYcoherence=0
        nearest_bird, distance = predator.find_nearest_bird(birdList)
        if distance < 5:
            predator.maxSpeed = 2.5
        
                
            # coherence: move towards nearst bird
        forceXcoherence = (nearest_bird.x - predator.x)
        forceYcoherence = (nearest_bird.y - predator.y)

        distances = {
            'left': predator.x ,
            'right': (canvas.winfo_width() - predator.x),
            'top': predator.y,
            'bottom': (canvas.winfo_height() - predator.y)
        }
        handleEdgeX=0
        handleEdgeY=0
        for direction, distance in distances.items():
            if distance < predator.edge_buffer:
                strength = ((predator.edge_buffer - distance) / (predator.edge_buffer)) 
                if direction == 'left':
                    handleEdgeX = strength 
                elif direction == 'right':
                    handleEdgeX = -strength 
                elif direction == 'top':
                    handleEdgeY = strength  
                elif direction == 'bottom':
                    handleEdgeY = -strength 

        predator.ax = 0.0
        predator.ax += handleEdgeX*100
        predator.ax += forceXcoherence*0.001




        predator.ay = 0.0
        predator.ay += handleEdgeY*100
        predator.ay += forceYcoherence*0.001
    return canvas,listOfPredators






def moveBirds(canvas,listOfBirds,islands,predatorsList):
    global food_position
    global foods
    global died_birds
    new_died_birds=[]
    for bird in listOfBirds:
        for otherP in predatorsList:
            dp = bird.distance(otherP)
            if dp < 5:
                bird.alive = False
        if bird.alive==False:
            draw_red_x(canvas,bird.x,bird.y)
        bird.move(canvas)
    showNeighboursB(canvas,listOfBirds,islands,predatorsList)
    
    # Call Predator move here
    canvas,predatorsList = movePredator(predatorsList,canvas,listOfBirds,islands)
    
    for bird in listOfBirds:
        # finding neighbouring birds
        nearbyBirds = []
        for otherBird in listOfBirds:
            d = bird.distance(otherBird)
            if not otherBird == bird and d < bird.radius and d!=0:
                nearbyBirds.append(otherBird)
            
        
        # alignment: orient yourself towards the neighbours
        averageVx = 0.0
        averageVy = 0.0
        if nearbyBirds:
            for nb in nearbyBirds:
                
                averageVx += nb.vx
                averageVy += nb.vy
            averageVx /= len(nearbyBirds)
            averageVy /= len(nearbyBirds)
        forceXalignment = averageVx - bird.vx
        forceYalignment = averageVy - bird.vy
        
        # coherence: move towards average position of neighbours
        averageX = 0.0
        averageY= 0.0
        if nearbyBirds:
            for nb in nearbyBirds:
                averageX +=  nb.x
                averageY +=  nb.y
            averageX /= len(nearbyBirds)
            averageY /= len(nearbyBirds)
        forceXcoherence = (averageX - bird.x)
        forceYcoherence = (averageY - bird.y)
        
        # separation: move away proportional to neighbour distance
        averageX = 0.0
        averageY = 0.0
        if nearbyBirds:
            for nb in nearbyBirds:
                differenceX = bird.x - nb.x
                differenceY = bird.y - nb.y
                differenceX /= bird.distance(nb)
                differenceY /= bird.distance(nb)
                averageX += differenceX
                averageY += differenceY
            averageX /= len(nearbyBirds)
            averageY /= len(nearbyBirds)
        forceXseparation = averageX
        forceYseparation = averageY
        
        distances = {
            'left': bird.x ,
            'right': (canvas.winfo_width() - bird.x),
            'top': bird.y,
            'bottom': (canvas.winfo_height() - bird.y)
        }
        handleEdgeX=0
        handleEdgeY=0
        for direction, distance in distances.items():
            if distance < bird.edge_buffer:
                strength = ((bird.edge_buffer - distance) / (bird.edge_buffer))
                if direction == 'left':
                    handleEdgeX = strength  
                elif direction == 'right':
                    handleEdgeX = -strength  
                elif direction == 'top':
                    handleEdgeY = strength  
                elif direction == 'bottom':
                    handleEdgeY = -strength 

        # separation: move away proportional to predators distance
        for predator in predatorsList:
            nearbyPredators = []
            for otherPredator in predatorsList:
                d = bird.distance(otherPredator)
                if d < bird.radius:
                    nearbyPredators.append(otherPredator)

        
        averageX = 0.0
        averageY = 0.0
        if nearbyPredators:
            for np in nearbyPredators:
                differenceX = bird.x - np.x
                differenceY = bird.y - np.y
                differenceX /= bird.distance(np)
                differenceY /= bird.distance(np)
                averageX += differenceX
                averageY += differenceY
            averageX /= len(nearbyPredators)
            averageY /= len(nearbyPredators)
        forceXseparationP = averageX
        forceYseparationP = averageY


        #====chase food
        forceFoodX=0
        forceFoodY=0
        closest_food = None
        min_distance = 100000
        for (fx, fy, _) in food_position:
            distance = math.sqrt((bird.x - fx)**2 + (bird.y - fy)**2)
            if distance < min_distance:
                min_distance = distance
                closest_food = (fx, fy)
        if closest_food:
            forceFoodX = (closest_food[0] - bird.x)   
            forceFoodY = (closest_food[1] - bird.y)
        
        to_remove = []
        for (fx, fy, fid) in food_position:
            if math.sqrt((bird.x - fx)**2 + (bird.y - fy)**2) < 10: 
                bird.maxSpeed = 2.5
                bird.hungry += 100
                canvas.delete(fid)  
                to_remove.append((fx, fy, fid))
        
        for item in to_remove:
            food_position.remove(item)
            foods-= 1
            
        
        
        # total influence
        
        bird.ax = 0.0
        bird.ax += forceXalignment*0.005
        # bird.ax += forceXcoherence*0.55
        bird.ax += forceXseparation*800
        bird.ax += handleEdgeX*69999
        bird.ax += forceXseparationP*9999
        bird.ax += forceFoodX*60


        bird.ay = 0.0
        bird.ay += forceYalignment*0.005
        # bird.ay += forceYcoherence*0.55
        bird.ay += forceYseparation*800
        bird.ay += handleEdgeY*69999
        bird.ay += forceYseparationP*9999
        bird.ay += forceFoodY*60

    return listOfBirds,canvas,islands ,predatorsList



        
       

def showNeighboursP(canvas,listOfBirds,islands,listOfPredator):
    for p in listOfPredator:
        radius = p.radius
        canvas.delete("circle")
        canvas.create_oval(p.x-radius/3, p.y-radius/3,\
                           p.x+radius/3, p.y+radius/3, tags="circle")
    
def showNeighboursB(canvas,listOfBirds,islands,predatorsList):
    for p in predatorsList:
        nearbyBirds = []
        nearestBird = None
        radius= p.radius
        for otherBird in listOfBirds:
            if p.distance(otherBird)<=radius/3:
                nearbyBirds.append(otherBird)
        if nearbyBirds:
            shortestDistance = float('inf')
            for otherBird in nearbyBirds:
                d = p.distance(otherBird)
                if d < shortestDistance:
                    shortestDistance = d
                    nearestBird = otherBird
        p.color = "black"
        for n in nearbyBirds:
            n.color = "red"
        if not nearestBird==None:
            nearestBird.color = "cyan"
        for bird in listOfBirds:
            bird.draw(canvas)
        for b in listOfBirds:
            b.color = "blue"
            
    for otherBird in listOfBirds:
        otherBird.on_land = False  
        for island in islands:
            if is_point_in_polygon((otherBird.x, otherBird.y), island.points):
                otherBird.color = "green"  
                otherBird.on_land = True
                break
            if not otherBird.on_land:
                otherBird.color = "blue"  

    
        




```

```python
from stable_baselines3 import PPO

def train_model(env):
    model = PPO("MlpPolicy", env, verbose=1)
    model.learn(total_timesteps=99999)
    model.save("bird_model")
    print("Model training complete and saved.")


import  gym
from gym import spaces
import numpy as np

class MultiBirdEnv(gym.Env):
    def __init__(self,canvas):
        super(MultiBirdEnv, self).__init__()
        self.canvas = canvas
        self.birds = addBirds(self.canvas) 
        self.predators = addPredators(self.canvas)
        self.islands = addIsland(self.canvas)
        self.num_birds = len(self.birds)
        self.died=0
        self.current_step = 0
        # The action of each bird is to control ax and ay
        self.action_space = spaces.MultiDiscrete([3] * self.num_birds * 2)
        global food_position
        self.max_food=20
        self.max_birds=60
        self.max_predators=1

       # The observations are the location of the bird, food, and predator
        num_features_per_bird = 2  
        num_features_per_food = 2  
        num_features_per_predator = 2  
        max_features = (60 * num_features_per_bird +
                          20 * num_features_per_food +
                          1 * num_features_per_predator)
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(max_features,), dtype=np.float32)


    def reset(self):
        self.current_step = 0
        self.canvas.delete('all')
        # Re-create and initialize canvas
        self.birds = addBirds(self.canvas)  
        self.predators = addPredators(self.canvas)
        self.islands = addIsland(self.canvas)
        return self._get_observation()

    def step(self, actions):
        rewards = 0
        self.current_step += 1
        global food_position
        self.birds,self.canvas,self.islands ,self.predators = moveBirds(self.canvas,self.birds,self.islands,self.predators)
        ax_action = (actions % 3) - 1  # Convert from 0,1,2 to -1,0,1
        ay_action = (actions // 3) - 1  
        done = self._check_done()
        for idx, bird in enumerate(self.birds):
            ax_index = idx * 2    # Calculate the index of the ax action
            ay_index = ax_index + 1 
            ax_action = actions[ax_index] - 1  # Convert to -1, 0, 1
            ay_action = actions[ay_index] - 1 
            bird.ax += ax_action*9
            bird.ay += ay_action*9
            bird.move(self.canvas)
            rewards += self._calculate_reward(bird,self.birds,self.predators,food_position)
        # self.window.mainloop()  not working
        return self._get_observation(), rewards, done, {}

    def _get_observation(self):
        # # Collect all status information
        padding_value = np.nan
        global food_position
        bird_data = np.array([[bird.x, bird.y] for bird in self.birds] )
        food_data = np.array([[food[0], food[1]] for food in food_position] + [[padding_value, padding_value]] * (self.max_food - len(food_position)))
        predator_data = np.array([[predator.x, predator.y] for predator in self.predators])
        # velocity_data = np.array([[bird.vx, bird.vy] for bird in self.birds] )

        # Only non-empty arrays are connected
        observation = np.concatenate([bird_data.flatten(), food_data.flatten(), predator_data.flatten()])
        observation = np.nan_to_num(observation, nan=-999)

        return observation
        
    def _calculate_reward(self,bird,birdList,predatorsList,food_position):
        reward = 0
        nearbyBirds = []
        for otherBird in birdList:
            d = bird.distance(otherBird)
            if not otherBird == bird and d < bird.radius/2 and d!=0:
                nearbyBirds.append(otherBird)
                # if d < bird.radius/3 :
                    # reward -= 5
        reward += len(nearbyBirds)*10
        reward -= 2

        
        

                
        return reward

    def _check_done(self):
        if self.current_step > 3000: 
            return True
        else: 
            return False

import tkinter as tk
import threading
def main():
    window = tk.Tk()
    canvas = initialise(window)
    env = MultiBirdEnv(canvas)  # Create environment
    model = PPO("MlpPolicy", env, verbose=1, learning_rate=0.0025, n_steps=2000, gamma=0.1, batch_size=1000)
    model.learn(total_timesteps=50000)
    # training
    model.save("bird_model_ppo")
    print("Model training complete and saved.")

main()
```
